```python
#import libraries and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

df = pd.read_csv('cleanmath.csv')
df
```

```python
ax = sns.boxplot(x='X2MTHID', y='X2TXMQUINT', data=df)
#Math identity and math scores showed a 0.4 correlation. 
#A boxplot was used to further explore this relationship.
```

```python
#Group 1: High identity, low scores = Over confident ---> 
#created a subset of 204 falsely confident folks
df_OC = df[(df['X2MTHID']>=1.28) & (df['X2TXMQUINT'] <= 2)]
df_OC

#H1: My hypothesis is that there will be more boys in group 1 (overconfident).
```

```python
#Group 2: High scores, low identity = Under confident ---> created a subset of 187 not confident folks
df_UC = df[(df['X2MTHID'] <= -1.5) & (df['X2TXMQUINT'] == 5)]
df_UC
#H2: My hypothesis is that there will be more girls in group 2 (underrconfident).
```

```python
#Group 3: High scores and high identity ----> created a subset of 1012 high performers/believers.
df_HP = df[(df['X2MTHID']>1.8) & (df['X2TXMQUINT'] == 5)]
df_HP
#H3: My hypothesis is that there will be an equal number of boys and girls in group 3 (high perfomers/believers).
```
